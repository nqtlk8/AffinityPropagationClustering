# Data Overview

1. Import data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.cluster import AffinityPropagation
import warnings
```

```python
data = pd.read_csv('CC GENERAL.csv')
df = pd.DataFrame(data)
df
```

2. Data Overview

```python
df.describe()
```

3. Remove missing data

```python
df = df.dropna()
df.isnull().sum()
```

4. Select Feature and Samples

```python
# Chỉ lấy các cột quan trọng có ý nghĩa 
features = ['BALANCE','CREDIT_LIMIT','PURCHASES','PURCHASES_FREQUENCY']
df_subset = df[features].dropna().copy()
df_subset
```

```python
# Vì thuật toán AP có độ phức tạp cao O(N*N) nên mình sẽ chỉ lấy samples để đại diện dataset
# Tạo một cột tạm thời để "phân tầng" (Stratify)
# Chia CREDIT_LIMIT thành 10 nhóm (deciles) có số lượng xấp xỉ nhau
df_subset['Strata_Limit'] = pd.qcut(df_subset['CREDIT_LIMIT'], q=10, labels=False, duplicates='drop')

# Lấy mẫu phân tầng đúng 1000 dòng
# train_size=1000 giúp lấy đúng số lượng, stratify đảm bảo tỷ lệ các nhóm được giữ nguyên
df_samples, _ = train_test_split(
    df_subset, 
    train_size=1000, 
    stratify=df_subset['Strata_Limit'], 
    random_state=42 # Set seed để kết quả ổn định qua các lần chạy
)

# Xóa cột phân tầng tạm thời vì không còn cần thiết
df_samples = df_samples.drop(columns=['Strata_Limit'])
df_subset = df_subset.drop(columns=['Strata_Limit'])

# Vẽ biểu đồ Scatter Plot để so sánh hình dáng dữ liệu
# Chọn 2 trục là BALANCE và PURCHASES để quan sát rõ nhất các hành vi
plt.figure(figsize=(14, 6))

# Biểu đồ 1: Dữ liệu gốc (Toàn bộ df_subset)
plt.subplot(1, 2, 1)
sns.scatterplot(
    data=df_subset, 
    x='BALANCE', 
    y='PURCHASES', 
    alpha=0.3, # Làm trong suốt các điểm để thấy rõ mật độ
    color='blue'
)
plt.title(f'Tập dữ liệu gốc ({len(df_subset)} dòng)')
plt.xlabel('Balance (Số dư)')
plt.ylabel('Purchases (Chi tiêu)')

# Biểu đồ 2: Tập mẫu phân tầng (df_samples)
plt.subplot(1, 2, 2)
sns.scatterplot(
    data=df_samples, 
    x='BALANCE', 
    y='PURCHASES', 
    alpha=0.6, 
    color='red'
)
plt.title(f'Tập lấy mẫu phân tầng ({len(df_samples)} dòng)')
plt.xlabel('Balance (Số dư)')
plt.ylabel('Purchases (Chi tiêu)')

plt.tight_layout()
plt.show()
 
```

# K-Mean Clustering

1. Nomarlization

```python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df_samples)
```

```python
wcss = []            # Lưu trữ tổng bình phương khoảng cách (Inertia) cho Elbow
silhouette_avg = []  # Lưu trữ điểm Silhouette trung bình
k_values = range(2, 11) 

for k in k_values:
    # Khởi tạo và huấn luyện mô hình K-Means
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42, n_init=10)
    kmeans.fit(scaled_features)
    
    # Tính WCSS cho phương pháp Elbow
    wcss.append(kmeans.inertia_)
    
    # Tính Silhouette Score
    cluster_labels = kmeans.labels_
    silhouette_avg.append(silhouette_score(scaled_features, cluster_labels))

# 3. TRỰC QUAN HÓA HAI PHƯƠNG PHÁP
fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# Biểu đồ 1: Elbow Method (Phương pháp điểm gãy)
axes[0].plot(k_values, wcss, marker='o', linestyle='--', color='b')
axes[0].set_title('Elbow Method (Xác định điểm gãy)')
axes[0].set_xlabel('Số lượng cụm (k)')
axes[0].set_ylabel('WCSS (Inertia)')
axes[0].grid(True)

# Biểu đồ 2: Silhouette Score
axes[1].plot(k_values, silhouette_avg, marker='s', linestyle='-', color='g')
axes[1].set_title('Silhouette Score (Hệ số Silhouette)')
axes[1].set_xlabel('Số lượng cụm (k)')
axes[1].set_ylabel('Silhouette Score')
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

Theo Chart, tại k=4 là tối ưu và phù hợp và bài toán

```python
# Visual Clustering
# 1. Chạy lại K-Means với k=4 để gán nhãn cụm
OPTIMAL_K = 4
kmeans = KMeans(n_clusters=OPTIMAL_K, init='k-means++', random_state=42, n_init=10)

# (Giả định scaled_features đã được tạo ở bước trước, nếu chưa thì bỏ comment dòng dưới)
# scaled_features = StandardScaler().fit_transform(df_samples[['BALANCE', 'CREDIT_LIMIT', 'PURCHASES', 'PURCHASES_FREQUENCY']])

df_samples['Cluster'] = kmeans.fit_predict(scaled_features)

# Ép kiểu cột Cluster sang dạng chuỗi (String/Category) để đồ thị hiển thị màu sắc rời rạc (Categorical)
df_samples['Cluster'] = df_samples['Cluster'].astype(str)

# 2. Trực quan hóa kết quả phân cụm
sns.set_theme(style="whitegrid")
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Biểu đồ 1: Quan sát quy mô tiền tệ (Số dư vs Tổng chi tiêu)
sns.scatterplot(
    data=df_samples, 
    x='BALANCE', 
    y='PURCHASES', 
    hue='Cluster', 
    palette='Set1', # Sử dụng bộ màu tương phản cao
    alpha=0.7,
    s=50, # Kích thước điểm
    ax=axes[0]
)
axes[0].set_title('Quy mô tài chính: Số dư (Balance) vs Chi tiêu (Purchases)', fontsize=14, pad=15)
axes[0].set_xlabel('Balance (Số dư nợ)', fontsize=12)
axes[0].set_ylabel('Purchases (Tổng chi tiêu)', fontsize=12)

# Biểu đồ 2: Quan sát thói quen (Hạn mức vs Tần suất mua sắm)
sns.scatterplot(
    data=df_samples, 
    x='CREDIT_LIMIT', 
    y='PURCHASES_FREQUENCY', 
    hue='Cluster', 
    palette='Set1',
    alpha=0.7,
    s=50,
    ax=axes[1]
)
axes[1].set_title('Thói quen: Hạn mức (Credit Limit) vs Tần suất (Frequency)', fontsize=14, pad=15)
axes[1].set_xlabel('Credit Limit (Hạn mức)', fontsize=12)
axes[1].set_ylabel('Purchases Frequency (Tần suất mua sắm, 0-1)', fontsize=12)

plt.tight_layout()
plt.show()
```

# Afinity Propagation Clustering

```python
# 1. Tạo danh sách các giá trị preference để thử nghiệm
# Vì dữ liệu đã được StandardScaler (trung bình 0, phương sai 1), khoảng cách giữa các điểm không quá lớn.
# Ta sẽ thử nghiệm từ một số âm lớn (ít cụm) tiến dần về 0 (nhiều cụm).
# Quét từ -400 tiến dần lên -50, mỗi bước nhảy là 20
preferences_to_test = np.arange(-400, -100, 20)

# Biến để lưu trữ kết quả tốt nhất
best_score = -1
best_preference = None
best_n_clusters = 0
best_labels = None
best_exemplars_indices = None

print("ĐANG TÌM KIẾM PREFERENCE TỐI ƯU CHO AFFINITY PROPAGATION...\n")
print(f"{'Preference':<15} | {'Số lượng cụm':<15} | {'Silhouette Score':<15}")
print("-" * 55)

# 2. Chạy vòng lặp thử nghiệm
for pref in preferences_to_test:
    # Tắt cảnh báo hội tụ để màn hình console sạch sẽ
    with warnings.catch_warnings():
        warnings.simplefilter("ignore")
        
        # Khởi tạo và fit mô hình với preference hiện tại
        ap = AffinityPropagation(preference=pref, random_state=42)
        ap.fit(scaled_features)
        
    n_clusters_ = len(ap.cluster_centers_indices_)
    
    # Chỉ tính Silhouette Score khi có từ 2 cụm trở lên
    if n_clusters_ > 1:
        score = silhouette_score(scaled_features, ap.labels_)
        print(f"{pref:<15} | {n_clusters_:<15} | {score:.4f}")
        
        # So sánh và cập nhật nếu tìm thấy điểm số tốt hơn
        if score > best_score:
            best_score = score
            best_preference = pref
            best_n_clusters = n_clusters_
            best_labels = ap.labels_
            best_exemplars_indices = ap.cluster_centers_indices_
    else:
        print(f"{pref:<15} | {n_clusters_:<15} | Không đủ cụm (N/A)")

# 3. In ra kết quả tốt nhất cuối cùng
print("\n" + "=" * 55)
print(" TỔNG KẾT KẾT QUẢ TỐT NHẤT")
print(f" Preference tối ưu: {best_preference}")
print(f" Số lượng cụm (k): {best_n_clusters}")
print(f" Silhouette Score: {best_score:.4f}")
print("=" * 55)

# 4. Hiển thị chân dung các Exemplars của kết quả tốt nhất
if best_n_clusters > 0:
    print(f"\n--- CHÂN DUNG {best_n_clusters} KHÁCH HÀNG ĐẠI DIỆN (EXEMPLARS) ---")
    best_exemplars_df = df_samples.iloc[best_exemplars_indices]
    print(best_exemplars_df[['BALANCE', 'CREDIT_LIMIT', 'PURCHASES', 'PURCHASES_FREQUENCY']])
```

```python
# Visual Clsuter

# 1. Áp dụng AP với preference tối ưu
ap_final = AffinityPropagation(preference=-380, random_state=42)

# Chạy mô hình và lưu nhãn cụm vào dataframe
df_samples['AP_Cluster'] = ap_final.fit_predict(scaled_features)

# Lấy index của các tâm cụm (Exemplars)
exemplars_indices = ap_final.cluster_centers_indices_

# Trích xuất dữ liệu của các tâm cụm trên TẬP GỐC (chưa chuẩn hóa) để vẽ đúng tọa độ
exemplars_df = df_samples.iloc[exemplars_indices]

# Ép kiểu biến Cluster sang string để vẽ màu sắc phân loại rời rạc
df_samples['AP_Cluster'] = df_samples['AP_Cluster'].astype(str)

# 2. Trực quan hóa dữ liệu và Tâm cụm
sns.set_theme(style="whitegrid")
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# --- Biểu đồ 1: BALANCE vs PURCHASES ---
sns.scatterplot(
    data=df_samples, x='BALANCE', y='PURCHASES', 
    hue='AP_Cluster', palette='Set2', alpha=0.6, s=50, ax=axes[0], legend=False
)
# Vẽ tâm cụm (Exemplars) đè lên đồ thị
axes[0].scatter(
    exemplars_df['BALANCE'], exemplars_df['PURCHASES'], 
    color='gold', marker='*', s=400, edgecolor='black', linewidth=1.5, zorder=10
)
axes[0].set_title('Số dư vs Chi tiêu (Tâm cụm là Ngôi sao)', fontsize=14, pad=15)
axes[0].set_xlabel('Balance (Số dư nợ)')
axes[0].set_ylabel('Purchases (Tổng chi tiêu)')

# --- Biểu đồ 2: CREDIT_LIMIT vs PURCHASES_FREQUENCY ---
sns.scatterplot(
    data=df_samples, x='CREDIT_LIMIT', y='PURCHASES_FREQUENCY', 
    hue='AP_Cluster', palette='Set2', alpha=0.6, s=50, ax=axes[1]
)
# Vẽ tâm cụm (Exemplars) đè lên đồ thị
axes[1].scatter(
    exemplars_df['CREDIT_LIMIT'], exemplars_df['PURCHASES_FREQUENCY'], 
    color='gold', marker='*', s=400, edgecolor='black', linewidth=1.5, zorder=10
)
axes[1].set_title('Hạn mức vs Tần suất (Tâm cụm là Ngôi sao)', fontsize=14, pad=15)
axes[1].set_xlabel('Credit Limit (Hạn mức)')
axes[1].set_ylabel('Purchases Frequency (Tần suất)')

plt.tight_layout()
plt.show()
```
